Problema 1
-----------

*Dos vehículos de competición inician una carrera con la misma aceleración $\vec{a_i}=(3 + t)\vec{i}$ (medida en $m/s^{2}$ y $t$ en segundos). Pasados 6 segundos, el segundo coche se estropea y, por efecto de la fricción por el aire, pasa a tener una aceleración $\vec{a_2} = -kv^2\vec{i}$(donde $k=0.01m^{-1}$ es constante). Determinar la distancia que aventajará el primero al segundo 12 segundos después de empezar la carrera.*

Para el primer coche partimos de la información que nos dan: aceleración $\vec{a_i}=3 + t$, e inicio en el origen desde el reposo $(v_0 = 0, x_0 = 0)$ para ambos coches. Por simplicidad trabajaremos siempre en la componente $\vec{i}$ reduciendo el problema a una dimensión.

En primer lugar calculamos las ecuaciones de movimiento del primer coche. A partir de la aceleración podemos integrar para obtener la velocidad (entre $v_0 = 0$ y $v_1(t)$) y a partir de ésta, la posición (entre $x_0 = 0$ y $x_1(t)$):

$$a_1(t) = t + 3 \frac{m}{s^{2}}$$
Obtenemos $v_1(t)$:
$$\frac{dv_1(t)}{dt} = t + 3,$$
$$dv_1(t) = (t + 3)dt,$$
$$\int_{v_0}^{v(t)}dv_1 = \int_0^t(t + 3)dt,$$
$$v_1(t) - v_0 = \frac{t^2}{2} + 3t,$$
$$v_1(t) = \frac{t^{2}}{2} + 3t + v_0$$
Obtenemos $x_1(t)$:
$$\frac{dx_1(t)}{dt} = \frac{t^{2}}{2} + 3t + v_0,$$
$$dx_1(t) = \left(\frac{t^{2}}{2} + 3t + v_0\right)dt,$$
$$\int_{x_0}^{x(t)}dx_1 = \int_0^t\left(\frac{t^{2}}{2} + 3t + v_0\right)dt,$$
$$x_1(t) - x_0 = \frac{t^3}{6} + 3\frac{t^2}{2} + tv_0,$$
$$x_1(t) = \frac{t^3}{6} + 3\frac{t^2}{2} + tv_0 + x_0$$
Como parten del origen en reposo $x_0 = 0$ y $v_0 = 0$, tenemos que las ecuaciones de movimiento son:
$$a_1(t) = t + 3 \frac{m}{s^{2}},$$
$$v_1(t) = \frac{t^{2}}{2} + 3t \frac{m}{s},$$
$$x_1(t) = \frac{t^{3}}{6} + \frac{3t^2}{2} m$$

En el caso del segundo vehículo la complejidad aumenta; suponemos que $v_2(t) = v_1(t) \forall t \in [0, 6]$ y procedemos a calcular las ecuaciones de movimiento del segundo coche para el intervalo $t \in (6, 12]$.

$$a_2(t) = -k(v_2(t))^2\frac{m}{s^{2}}$$
Obtenemos $v_2(t)$:
$$\frac{dv_2(t)}{dt} = -k(v_2(t))^2,$$
$$\frac{dv_2(t)}{(v_2(t))^2}=-kdt,$$
Como en $t=6$, $v_1(6) = v_2(6)$, usamos $v_6 = v_1(6) = v_2(6)$:
$$\int^{v_2(t)}_{v_6}\frac{dv}{v^2} = -k\int_6^{t}dt,$$
$$\frac{-1}{v}\Big|_{v_6}^{v_2(t)} = -kt\Big|^t_6,$$
$$-\left(\frac{1}{v_2(t)}-\frac{1}{v_6}\right) = -k(t - 6),$$
$$\frac{1}{v_2(t)} = k(t - 6) + \frac{1}{v_6},$$
$$v_2(t) = \frac{1}{k(t - 6) + \frac{1}{v_6}},$$
Obtenemos $x_2(t)$:
$$\frac{dx_2(t)}{dt} = \frac{1}{k(t - 6) + \frac{1}{v_6}},$$
$$dx_2(t) = \frac{dt}{k(t - 6) + \frac{1}{v_6}},$$
Como en $t=6$, $x_1(6) = x_2(6)$, usamos $x_6 = x_1(6) = x_2(6)$:
$$\int^{x_2(t)}_{x_6}dx_2 = \int_6^t\frac{dt}{k(t - 6) + \frac{1}{v_6}},$$
Nos compensa un cambio de variable; si $u(t) = k(t - 6) + \frac{1}{v_6}$ su derivada con respecto a $t$ será $du = kdt$ por lo que $dt = \frac{du}{k}$ y el valor de $u$ en $t = 6$ quedaría $u(6) = \frac{1}{v_6}$, sustituyendo en la ecuación anterior tenemos que:
$$\int^{x_2(t)}_{x_6}dx_2 = \frac{1}{k}\int_{u(6)}^{u(t)}\frac{du}{u},$$
$$x_2\Big|^{x_2(t)}_{x_6} = \frac{1}{k}\ln(u)\Big|_{u(6)}^{u(t)},$$
$$x_2(t) - x_6 = \frac{1}{k}(\ln(u(t)) - \ln(u(6))),$$
$$x_2(t) = \frac{1}{k}\ln\left(\frac{u(t)}{u(6)}\right) + x_6,$$
Deshacemos el cambio de variable,
$$x_2(t) = \frac{1}{k}\ln\left(\frac{k(t - 6) + \frac{1}{v_6}}{\frac{1}{v_6}}\right) + x_6,$$
$$x_2(t) = \frac{\ln(1 + kv_6(t - 6))}{k} + x_6$$

Por lo que las ecuaciones de movimiento del segundo coche para el intervalo $t \in (6, 12]$ quedan:
$$a_2(t) = -k(v_2(t))^2\frac{m}{s^{2}},$$
$$v_2(t) = \frac{1}{k(t - 6) + \frac{1}{v_6}}\frac{m}{s},$$
$$x_2(t) = \frac{\ln(1 + kv_6(t - 6))}{k} + x_6 m$$




```python
import matplotlib.pyplot as plt
import numpy as np

# Global
t = np.linspace(0, 12, 100)
t1 = np.linspace(0, 6, 50)
t2 = np.linspace(6.01, 12, 50)
k = 0.01 # 1 / m

def a1(t):    
    return 3 + t

def v1(t):
    return t ** 2 * 0.5 + 3 * t

def x1(t):
    return t ** 3 / 6. + t ** 2 * 3 / 2.

def x2(t):
    if t > 6:
        return np.log(1 + k * v1(6) * (t - 6)) / k + x1(6)
    else:
        return x1(t)        

def v2(t):
    if t > 6:
        return 1 / (k * (t - 6) + 1 / v1(6))
    else:
        return v1(t)        

def a2(t):
    if t > 6:
        return -k * v2(t) ** 2        
    else:
        return a1(t)        

fig = plt.figure()
fig.suptitle('Aceleracion', fontsize=10, fontweight='bold')
ax = fig.add_subplot(111)
ax.grid(linestyle='--')
ax.set_xlabel('t[s]')
ax.set_ylabel('a[m/s^2]')
ax.plot(t, map(a1, t), label='a1', color='blue')
ax.plot(t1, map(a2, t1), label='a2', color='orange')
ax.plot(t2, map(a2, t2), color='orange')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
fig.suptitle('Velocidad', fontsize=10, fontweight='bold')
ax = fig.add_subplot(111)
ax.grid(linestyle='--')
ax.set_xlabel('t[s]')
ax.set_ylabel('v[m/s]')
plt.plot(t, map(v1, t), label='v1')
plt.plot(t, map(v2, t), label='v2')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
fig.suptitle('Posicion', fontsize=10, fontweight='bold')
ax = fig.add_subplot(111)
ax.grid(linestyle='--')
ax.set_xlabel('t[s]')
ax.set_ylabel('x[m]')
plt.plot(t, map(x1, t), label='x1', color=)
plt.plot(t, map(x2, t), label='x2')
ax.legend()
plt.show()
```
