*Una noria de radio $R=10m$ gira con velocidad angular constante $\omega$ tal que cada cabina tarda 8 segundos en dar una vuelta completa. Cuando una de las cabinas se encuentra en el punto más alto, un tornillo de su base se suelta. Determinar, despreciando el tamaño de la cabina frente al de la noria:*

Del enunciado podemos recabar la siguente información:
$$f = \frac{1}{8}Hz,$$
$$\omega = 2\pi f = \frac{\pi}{4} \frac{rad}{s}$$

La única fuerza que está actuando sobre la cabina de la noria es la fuerza centrípeta y las ecuaciones de movimiento de la cabina son:
$$\vec{r}(t) = R\cos(\omega t)\vec{i} + R(1 + \sin(\omega t))\vec{j},$$
$$\vec{v}(t) = \frac{d\vec{r}(t)}{dt} = -R\omega\sin(\omega t)\vec{i} + R\omega\cos(\omega t)\vec{j},$$
$$\vec{a}(t) = \frac{d\vec{v}(t)}{dt} = -R\omega^2\cos(\omega t)\vec{i} - R\omega^2\sin(\omega t)\vec{j}$$

Cuando se desprende el tornillo tiene una velocidad inicial $\vec{v_0}$ equivalente a $\vec{v_0}(\frac{\pi}{2\omega}) = -R\omega\vec{i}$. (Por ser $\alpha = \omega t = \frac{\pi}{2}$ el punto más alto de la noria). Nótese que la componente de la velocidad sólo tiene contribución horizontal.

- *El tiempo que tarda el tornillo en alcanzar el suelo*

Como la contribución centrípeta de la noria a la velocidad inicial es únicamente horizontal no la consideramos para este apartado ($v_{y0} = 0 m/s$). Desde que el tornillo se desprende de la noria sufre una aceleración $g = 10 m/s^2$ debido a la acción de la gravedad. Este es un movimiento rectilíneo uniformemente acelerado que finalizará en cuanto el tornillo llegue al suelo ($y = 0$).

$$a_y(t) = -10 m/s^2,$$
$$v_y(t) = -10t + v_{y0} m/s,$$
$$y(t) = -5t^2 + v_{y0}t + y_0 m$$

Ya hemos razonado que $v_{y0}$. Ahora bien, $y_0$ será la posición inicial en el eje de abcisas del tornillo, que equivale al punto más alto de la noria $y_0 = 2*R = 20m$. Por lo que la  ecuación del movimiento del tornillo en el eje de abcisas queda:
$$y(t) = 20 - 5t^2$$
Buscamos el tiempo, $t$, que tarda en tocar el suelo, éste será el tiempo para el cual $y(t) = 0$:
$$20 - 5t^2 = 0,$$
$$\boxed{t = 2s}$$

- *La distancia del tornillo a la cabina en ese instante*
En este apartado nos interesa también calcular la distancia horizontal que ha recorrido del tornillo en esos 2 segundos. Así como lo que se ha desplazado la cabina en ese mismo tiempo. Una vez que tengamos las corrdenadas calculamos la distancia entre ambas.

Calculamos primero el desplazamiento de la cabina. Como $\omega = \frac{\pi}{4}\frac{rad}{s}$ el ángulo que habrá recorrido la noria en $2s$ será $\alpha(2) = \omega 2 = \frac{\pi}{2}$, por lo que habrá avanzado 90 grados y se situará en el extremo occidental de la noria. Las coordenadas de la cabina son por tanto:
$$\vec{r}(4) = 10\cos(\pi)\vec{i} + 10\vec{j},$$
$$\vec{r}(4) = -10\vec{i} + 10\vec{j} = 10(-1, 1)$$

Calculamos ahora las coordenadas del tornillo. Como está en el suelo savemos que $y = 0$. Para calcular la coordenada $x$, partimos de la velocidad inicial calculada anteriormente:
$$\vec{v_0}\left(\frac{\pi}{2\omega}\right) = -R\omega\vec{i},$$
$$\frac{dx(t)}{dt} = -R\omega,$$
$$x(t) = -R\omega t + x_0$$

Por cómo hemos situado los ejes tenemos que $x_0 = 0$, por tanto nos queda que $x(t) = -R\omega t$ y a los 2 segundos $x(2) = -5\pi m$. Por lo que tenemos:

 - Coordenadas cabina $r_{cab}= (-10, 10) m$
 - Coordenadas tornillo $r_{tor} = (-5\pi, 0) m$

La distancia entre ambos puntos es:
$$d = \sqrt{(-5\pi-10)^2 + (10)^2} = 11.51 m$$

- *La aceleración relativa del tornillo vista desde la cabina en el instante en el que se suelta*

Para calcular la aceleración relativa vista desde la cabina hacemos la diferencia entre la aceleración del tornillo $\vec{a_{tor}}(t) = -10 \vec{j} m/s^2$ y la aceleración de la cabina $\vec{a_{cab}} = \frac{d\vec{v}(t)}{dt} = -R\omega^2\cos(\omega t)\vec{i} - R\omega^2\sin(\omega t)\vec{j}$

$$\vec{a_{rel}}(t) = \frac{10\pi}{16}\cos\left(\frac{\pi t}{4}\right)\vec{i} + \left(\frac{10\pi}{16}\sin\left(\frac{\pi t}{4}\right) - 10\right)$$

Nótese que para alguien subido en la cabina tendrá la sensación de que el tornillo "sube" si la cabina baja más rápido que el tornillo (la cabina tiene una aceleración mayor que la del tornillo o dicho de otra manera $\frac{10\pi}{16}\sin\left(\frac{\pi t}{4}\right) > 10$). Cosa que nunca va a pasar porque la aceleración vertical máxima de la cabina es $a_{max} = \frac{10\pi}{16} = 1.96 m/s^2$ bastante inferior a los $g = 10 m / s^2$ de la gravedad.
