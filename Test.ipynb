Some equation $\frac{m}{s^{2}}$

```python
import matplotlib.pyplot as plt
import numpy as np

# Global
t = np.linspace(0, 12, 100)
k = 0.01 # 1 / m

def a1(t):    
    return 3 + t

def v1(t):
    return t ** 2 * 0.5 + 3 * t

def x1(t):
    return t ** 3 / 6. + t ** 2 * 3 / 2.

def x2(t):
    if t > 6:
        return (1 / k) * (np.log(k * (t - 6) + 1 / v1(6)) - np.log(1 / v1(6))) + x1(6)
    else:
        return x1(t)        

def v2(t):
    if t > 6:
        return 1 / (k * (t - 6) + 1 / v1(6))
    else:
        return v1(t)        

def a2(t):
    if t > 6:
        return -k * v2(t) ** 2        
    else:
        return a1(t)        

plt.plot(t, map(a1, t), label='a1')
plt.plot(t, map(a2, t), label='a2')
plt.legend()
plt.show()
```

```python
plt.plot(t, map(v1, t), label='v1')
plt.plot(t, map(v2, t), label='v2')
plt.legend()
plt.show()
```

```python
plt.plot(t, map(x1, t), label='x1')
plt.plot(t, map(x2, t), label='x2')
plt.legend()
plt.show()
```
