$$a_1(t) = t + 3 \frac{m}{s^{2}}$$
Obtenemos $v_1(t)$:
$$\frac{dv_1(t)}{dt} = t + 3,$$
$$dv_1(t) = (t + 3)dt,$$
$$\int_{v_0}^{v(t)}dv_1 = \int_0^t(t + 3)dt,$$
$$v_1(t) - v_0 = \frac{t^2}{2} + 3t,$$
$$v_1(t) = \frac{t^{2}}{2} + 3t + v_0$$
Obtenemos $x_1(t)$:
$$\frac{dx_1(t)}{dt} = \frac{t^{2}}{2} + 3t + v_0,$$
$$dx_1(t) = \left(\frac{t^{2}}{2} + 3t + v_0\right)dt,$$
$$\int_{x_0}^{x(t)}dx_1 = \int_0^t\left(\frac{t^{2}}{2} + 3t + v_0\right)dt,$$
$$x(t) - x_0 = \frac{t^3}{6} + 3\frac{t^2}{2} + tv_0,$$
$$x(t) = \frac{t^3}{6} + 3\frac{t^2}{2} + tv_0 - x_0$$
Como parten del origen en reposo $x_0 = 0$ y $v_0 = 0$, tenemos que las ecuaciones de movimiento son:
$$a_1(t) = t + 3 \frac{m}{s^{2}},$$
$$v_1(t) = \frac{t^{2}}{2} + 3t \frac{m}{s},$$
$$x_1(t) = \frac{t^{3}}{6} + \frac{3t^2}{2} m$$

```python
import matplotlib.pyplot as plt
import numpy as np

# Global
t = np.linspace(0, 12, 100)
k = 0.01 # 1 / m

def a1(t):    
    return 3 + t

def v1(t):
    return t ** 2 * 0.5 + 3 * t

def x1(t):
    return t ** 3 / 6. + t ** 2 * 3 / 2.

def x2(t):
    if t > 6:
        return (1 / k) * (np.log(k * (t - 6) + 1 / v1(6)) - np.log(1 / v1(6))) + x1(6)
    else:
        return x1(t)        

def v2(t):
    if t > 6:
        return 1 / (k * (t - 6) + 1 / v1(6))
    else:
        return v1(t)        

def a2(t):
    if t > 6:
        return -k * v2(t) ** 2        
    else:
        return a1(t)        

fig = plt.figure()
fig.suptitle('Aceleracion', fontsize=10, fontweight='bold')
ax = fig.add_subplot(111)
ax.grid(linestyle='--')
ax.set_xlabel('t[s]')
ax.set_ylabel('a[m/s^2]')
ax.plot(t, map(a1, t), label='a1')
ax.plot(t, map(a2, t), label='a2')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
fig.suptitle('Velocidad', fontsize=10, fontweight='bold')
ax = fig.add_subplot(111)
ax.grid(linestyle='--')
ax.set_xlabel('t[s]')
ax.set_ylabel('v[m/s]')
plt.plot(t, map(v1, t), label='v1')
plt.plot(t, map(v2, t), label='v2')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
fig.suptitle('Posicion', fontsize=10, fontweight='bold')
ax = fig.add_subplot(111)
ax.grid(linestyle='--')
ax.set_xlabel('t[s]')
ax.set_ylabel('x[m]')
plt.plot(t, map(x1, t), label='x1')
plt.plot(t, map(x2, t), label='x2')
ax.legend()
plt.show()
```
